판다스 내장 그래프 도구
- Matplotlib 라이브러리의 기능을 일부 내장
- DataFrame 객체.plot()
    - kind() 옵션 : 그래프의 종류 선택

선 그래프
- DataFrame 객체.plot(kind = "line")
    - kind = "line" 생략 가능 -> 기본 값
- 행 인덱스 -> x 축
- 열 이름 -> y 축

```python
#<예제3-4> 선 그래프 그리기

import pandas as pd

df = pd.read_excel("./남북한발전전력량.xlsx")

df_ns = df.iloc[[0, 5], 3:] # 남한, 북한 발전량 합계 데이터만 추출
df_ns.index = ["South","North"] # 행 인덱스 변경
df_ns.columns = df_ns.columns.map(int) # 열 이름의 자료형을 정수형으로 변경

print(df_ns.head())

print("\n")

# 선 그래프 그리기
print(df_ns.plot())

# 행 과 열을 바꾸어 변화 추이 보기 쉽게

tdf_ns= df_ns.T
tdf_ns.plot()
```

막대그래프
- DataFrame 객체.plot(kind = "bar")

```python
#<예제3-5> 막대 그래프 그리기

import pandas as pd

df = pd.read_excel("./남북한발전전력량.xlsx")

print(df.head())
print(df.iloc[[0,5],])

df_ns = df.iloc[[0,5],3:] # 남한, 북한의 합계만 사용
df_ns.index = ["South","North"] #  df_ns 의 행 인덱스 지정
df_ns.columns = df_ns.columns.map(int)  # 열 이름의 자료형을 정수형으로 변경

# 행 열 전치하여 새로운 데이터프레임 생성
tdf_ns = df_ns.T

# 막대 그래프 그리기
tdf_ns.plot(kind = "bar")
```

히스토그램
- DataFrame 객체.plot(kind = "hist")
- 히스토그램 = 도수 분포표

```python
#<예제 3-6> 히스토 그램

import pandas as pd

df = pd.read_excel("./남북한발전전력량.xlsx")

df_ns = df.iloc[[0,5],3:]
df_ns.index = ["South","North"]
df_ns.columns = df_ns.columns.map(int)

tdf_ns = df_ns.T

tdf_ns.plot(kind = "hist")
```

산점도
- DataFrame 객체.plot(x = "열 이름 1", y= "열 이름 2", kind = "scatter")
- 두 변수의 관계를 나타냄
- 데이터프레임의 열 중에서 비교할 두 열 을 선택하여 옵션에 넣는다.

```python
#<예제3-7> 삼전도

import pandas as pd

df = pd.read_csv("./auto-mpg.csv", header = None)

# 열 이름 지정
df.columns = ["mpg","cylinders","displacement","horsepower","weight","acceleration","model year", "origin","name"]

#  2 개의 열을 지정하여 산점도 그리기
df.plot(x ="weight", y = "mpg", kind = "scatter")
```

박스플롯
- Dataframe 객체[["열 이름1","열 이름2"]].plot("box")
- 특정 변수의 데이터 분포와 분산 정도를 나타냄

```python
#<예제3-8> 박스 플롯

import pandas as pd

df =pd.read_csv("./auto-mpg.csv", header = None)

df.columns = ["mpg","cylinders","displacement","horsepower","weight","acceleration","model year", "origin","name"]

# "mpg" 와 "cylinders" 에 대한 박스 플롯 그리기
df[["mpg","cylinders","displacement"]].plot(kind= "box")
```
