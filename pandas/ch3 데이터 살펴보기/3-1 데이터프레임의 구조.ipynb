데이터 내용 미리보기
- 앞부분 미리보기 : DataFrame객체.head(n) -> 디폴트 값 : 5
- 뒷부분 미리보기 : DataFrame객체.tail(n)

```python
#<예제3-1> 데이터 살펴보기

import pandas as pd

df = pd.read_csv("./auto-mpg.csv", header = None) # header = None 옵션 -> 열이름이 정수형으로 초기화

# 열 이름 지정
df.columns = ["mpg","cylinders","displacement","horsepower","weight","acceleration","model year","origin","name"]

# 데이터프레임의 일부분만 출력
print(df.head())

print("\n")

print(df.tail())
```

데이터 요약 정보 확인하기
- 데이터프레임의 크기
    - DataFrame 객체.shape
    - (행,열) 반환
    
    
- 데이터프레임의 기본 정보
    - DataFrame 객체.info()
    - 데이터프레임에 관한 기본 정보 반환
    
    
- 데이터프레임의 각 열의 자료형
    - DataFrame 객체.dtypes


- 데이터프레임의 기술 통계 정보 요약
    - DataFrame 객체.describe()
    - 산술(숫자)데이터에 대한
        - include = "all" 옵션 : 모든정보에 대한
    - 기술통계정보(평균, 표준편차, 최댓값, 최소값, 중간값, ...)

```python
#<예제3-1> 데이터 살펴보기

import pandas as pd

df = pd.read_csv("./auto-mpg.csv", header = None) # header = None 옵션 -> 열이름이 정수형으로 초기화

# 열 이름 지정
df.columns = ["mpg","cylinders","displacement","horsepower","weight","acceleration","model year","origin","name"]

#데이터프레임의 크기 출력
print(df.shape)

print("\n")

#데이터프레임의 기본 정보 출력
print(df.info())

print("\n")

# 데이터 자료형 츌력
print(df.dtypes)
print(type(df.dtypes))

print("\n")

# 데이터프레임의 기술통계 정보 출력
print(df.describe())
print("\n")
# 산술데이터가 아닌 열의 정보도 출력
print(df.describe(include = "all"))
```

데이터 개수 확인
- 각 열의 데이터 개수
    - DataFrame객체.count()
    - 유효한 값의 개수만을 계산
    

- 각 특정 열의 고유값 개수
    - DataFrame객체["열이름"].value_counts()
    - 고유값 개수를 세는데 사용
    - dropna =True옵션 : NaN 제외

```python
#<예제3-2> 데이터 개수확인

import pandas as pd

# read_csv로 df 객체 생성
df = pd.read_csv('./auto-mpg.csv',header = None) # header = None 옵션 : 열 을 하나 내려줌

# 열이름 지정
df.columns = ["mpg","cylinders","displacement","horsepower","weight","acceleration","model year","origin","name"]

# df의 각 열이 가지고있는 원소개수확인
print(df.count())

print("\n")

# df의 "origin" 열의 고유값 개수
# NaN 포함 : 디폴트값
print(df["origin"].value_counts())
print("\n")

# NaN 제외 : dropna = True 옵션
print(df["origin"].value_counts(dropna = True))
```
