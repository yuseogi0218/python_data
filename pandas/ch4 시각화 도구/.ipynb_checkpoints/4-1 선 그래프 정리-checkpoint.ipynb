그림판 객체 만들기 + 여러그래프 객체 만들기
fig = plt.figure(figsize = (가로크기, 세로크기))
axn(그래프 객체) = fig.add_subplot(전체 행, 전체 열, 순번)

하나의 그래프 그리기
plt(그래프객체).plot()

선 그래프 그리기
그래프객체.plot(x축 데이터, y축 데이터)
그래프객체.plot( 시리즈 또는 데이터프레임)
	-color, linewidth, ... 옵션 가능
	- marker(마커) 
		- "O" 만 표시 - 선 없이 마커만 표시
		- marker = "O" - 선 위에 마커 표시
		markerfacecolor, markersize, ...
	- 범례 지정
		- label = "범례 이름"
		그래프 밖에서 : 그래프객체.legend(loc = "best")

스타일 서식 지정
그래프 객체.style.use("서식")

그래프 제목 지정
그래프객체.title("제목이름")
	- 옵션 : size, color, weight, ...

축 이름 추가
	- x축 : 그래프객체.xlabel(" 축 이름")
	- y축 : 그래프객체.ylabe(" 축 이름")
		- 옵션 : size, color, ...

축 눈금 라벨 회전
	- x축 : 그래프객체.xticks()
	- y축 : 그래프객체.yticks()
		- 옵션 : rotation = 회전 각도 or "vertical)

축 눈금 라벨 지정 및 회전
	- x축: 그래프객체.set_xticklabels(라벨, rotation = )
	- y축: 그래프객체.set_yticklabels(라벨, rotation = )

축 눈금 라벨 크기
	- x축 : 그래프객체.tick_params(axis ="x", labelsize =크기)
	- y축 : 그래프객체.tick_params(axis ="y", labelsize =크기)

축 범위 지정
	- x축 : 그래프객체.set_xlim(최소값, 최대값)
	- y축 : 그래프객체.set_ylim(최소값, 최대값)

주석 사용
	- annotate() 함수 사용
		- 화살표
			- " ",
			xy = (x좌표, y좌표) # 화살표의 머리부분(끝점)
			xytext = (x좌표, y좌표) # 화살표의 꼬리부분(시작점)
			xycoords = "data) # 좌표체계
			arrowprops = dict(arrowstyle = "->",color = "색", lw = 5) # 화살표 서식

		- 텍스트
			- "텍스트" # 텍스트 입력
			xy= (x좌표, y좌표) # 텍스트 기준점
			rotation = # 회전각도
			va = "baseline" # 텍스트 상하정렬
			ha = "center" # 텍스트 좌우정렬
			fontsize = # 텍스트 크기
