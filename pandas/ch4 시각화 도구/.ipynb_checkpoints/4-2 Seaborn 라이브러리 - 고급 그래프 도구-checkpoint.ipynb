회귀선이 있는 산점도
- regplot() 함수
- 2개의 연속 변수 사이의 산점도
- 선형회귀분석에 의한 회귀선을 나타냄
    - fit_reg = 옵션 : 기본값 : True

```python
# <예제 4-26> 회귀선이 있는 산점도

import matplotlib.pyplot as plt # axe 객체활용을 위해
import seaborn as sns

titanic = sns.load_dataset("titanic")

#스타일 테마 설정(5가지 : darkgrid, whitegrid, dark, white, ticks)
sns.set_style("darkgrid")

# 그래프 객체 생성
fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

# 그래프그리기 = 선형 회귀선 표시(fit_reg = True :생략가능)
sns.regplot(x ="age", #x축 변수
           y = "fare", # y축 변수
           data =titanic, # 데이터
           ax = ax1) # axe객체 -1번째 그래프

# 그래프그리기 = 선형 회귀선 미표시(fit_reg = False)
sns.regplot(x ="age", #x축 변수
           y = "fare", # y축 변수
           data =titanic, # 데이터
           ax = ax2, # axe객체 -2번째 그래프
           fit_reg =False)

plt.show()
```

히스토그램 / 커널 밀도 그래프
- distplot() 함수
    - hist = 옵션 : True 기본값, False 표시하지 않음
    - kde(커널 밀도) = 옵션 : True 기본값, False 표시하지 않음

```python
# <예제 4-27> 히스토그램 / 커널 밀도 그래프

import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset("titanic")
sns.set_style("darkgrid")

# 그래프 객체 생성(figure에 3개의 서브 플롯 생성)
fig = plt.figure(figsize =(15,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

# 기본값 ( 히스토그램 / 커널 밀도그래프)
sns.distplot(titanic["fare"], ax = ax1)

# hist = False
sns.distplot(titanic["fare"], hist = False, ax = ax2)

# kde = False
sns.distplot(titanic["fare"], kde = False, ax =ax3)

# 차트 제목표시
ax1.set_title("titanic fare- hist/kde")
ax2.set_title("titanic fare- kde")
ax2.set_title("titanic fare- hist")

plt.show()
```

히트맵
- heatmap() 함수
    - annot = True 옵션 : 데이터 값 표시
    - fmt = "데이터 포맷"
    - cbar = False 옵션 : 컬러바 표시 X

```python
# <예제 4-28> 히트맵

import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset("titanic")
sns.set_style("darkgrid")

# 피벗테이블로 범주형 변수를 가각 행, 열로 재구분하여 정리
table = titanic.pivot_table(index =['sex'], columns = ['class'], aggfunc = 'size')

# 히트맵 그리기
sns.heatmap(table, # 데이터프레임
           annot = True, fmt = 'd', # 데이터 값 표시 여부, 정수형 포맷
           cmap = 'YlGnBu',
            linewidth =.5, # 구분선
           cbar = True) # 컬러 바 표시 여부

plt.show()
```

범주형데이터의 산점도
- stripplot() 함수
- swarmplot() 함수
    - 데이터의 분산까지 고려하야,데이터 포인트가 서로 중복 X
        - 데이터가 퍼져있는 정도를 볼 수 있음
        
- hue = "열 이름" 옵션 : 해당 열의 데이터를 가지고 구분

```python
# <예제 4-29> 범주형 데이터의 산점도

import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset("titanic")

sns.set_style("whitegrid")

# 그래프 객체 생성(figure에 2개의 서브플롯 생성)
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

#이산형 변수의 분포 - 데이터분산 미고려 - stripplot() 함수 사용
sns.stripplot(x = "class",
             y = "age",
             data = titanic,
             ax = ax1)

# 이산형 변수의 분포 - 데이터 분산 고려 - swarmplot() 함수 사용
sns.swarmplot(x = "class",
             y = "age",
             data = titanic,
              hue = "sex",# 남녀 성별을 색상으로 구분 가능
             ax = ax2)

# 차트 제목 표시
ax1.set_title("Strip plot")
ax2.set_title("Strip plot")

plt.show()
```

막대 그래프
- barplot() 함수 사용
    - hue = "열 이름" : 옵션 : 해당 열의 데이터를 가지고 구분
    - dodge = False 옵션 : stacke 그래프 그림 : 기본값 : True

```python
# <예제 4-30> 막대 그래프

import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset("titanic")

sns.set_style("whitegrid")

# 그래프 객체 새엉(figure에 3개의 서브플롯 생성)
fig = plt.figure(figsize =(15,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

# x축,y축에 변수 할당
sns.barplot(x = "sex", y ="survived",data = titanic, ax = ax1)

# x축,y축에 변수 할당하고 hue옵션 추가
sns.barplot(x = "sex", y = "survived", hue = "class", data = titanic, ax =ax2)

# x축, y축에 변수 할당하고, hue옵션을 추가하여 누적 출력
sns.barplot(x="sex",y="survived",hue ="class", dodge = False, data =titanic, ax =ax3)

#차트 제목 표시
ax1.set_title("titaic survived - sex")
ax1.set_title("titaic survived - sex/class")
ax1.set_title("titaic survived - sex/class(stacked)")

plt.show()
```

빈도 그래프
- countplot() 함수
    - hue = "열 이름" : 옵션 : 해당 열의 데이터를 가지고 구분
    - dodge = False 옵션 : stacke 그래프 그림 : 기본값 : True
- 각 범주에 속하는데이터의개수를 막대 그래프로 나타냄.

```python
# <예제 4-31>

import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset("titanic")

sns.set_style("whitegrid")

fig =plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

# 기본값
sns.countplot(x = "class", data = titanic, ax = ax1)

# hue 옵션에 "who" 추가
sns.countplot(x = "class", hue = "who", data = titanic, ax = ax2)

#dodge = False 옵션 추가 (누적 그래프)
sns.countplot(x = "class", hue = "who", dodge = False, data = titanic, ax = ax3)

# 차트 제목 표시
ax1.set_title("titanic class")
ax2.set_title("titanic class - who")
ax3.set_title("titanic class - who(stacked)")

plt.show()
```

박스 플롯 / 바이올린 그래프
- boxplot() / violinplot() 함수 사용
    - hue = "열 이름" : 옵션 : 해당 열의 데이터를 가지고 구분

```python
# <예제4-32> 박스 플롯 / 바이올린 그래프

import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset("titanic")

sns.set_style("whitegrid")

# 그래프 객체 생성(figure에 4개의 서브플롯 생성)
fig =plt.figure(figsize = (15,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

# 박스플롯 - 기본값
sns.boxplot(x = "alive", y = "age", data = titanic, ax = ax1)

# 박스플롯 - hue 변수 추가
sns.boxplot(x = "alive", y = "age", hue = "sex", data = titanic, ax = ax2)

# 바이올린 그래프 - 기본값
sns.violinplot(x = "alive", y = "age", data = titanic, ax = ax3)

# 바이올린 그래프 - hue 변수 추가
sns.violinplot(x = "alive", y = "age", hue = "sex", data = titanic, ax = ax4)

plt.show()
```

조인트 그래프
- 산점도 + x-y축에 각 변수에 대한 히스토그래프
- jointplot()함수 사용
    - kind = "reg": 회귀선 추가
    - kind = "hex" : 육각 산점도
    - kind = "kde" : 커널밀집 그래프

```python
# <예제 4-33> 조인트 그래프

import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset("titanic")

sns.set_style("whitegrid")

# 조인트 그래프 - 산점도(기본값)
j1 = sns.jointplot(x ="fare", y ="age", data=titanic)

# 조인트 그래프 - 회귀선
j2 = sns.jointplot(x = "fare", y= "age", kind = "reg", data =  titanic)

# 조인트 그래프 - 육각 산점도
j3 = sns.jointplot(x = "fare", y ="age", kind = "hex", data = titanic)

# 조인트 그래프 - 커널밀집 그래프
j4 = sns.jointplot(x = "fare", y= "age", kind = "kde", data = titanic)

# 차트 제목 표시
j1.fig.suptitle("titanicfare - scatter", size =15)
j2.fig.suptitle("titanicfare - reg", size =15)
j3.fig.suptitle("titanicfare - hex", size =15)
j4.fig.suptitle("titanicfare - kde", size =15)

plt.show()
```

조건을 적용하여 화면을 그리드로 분할하기
- 그래프 객체 = FacetGrid() 함수
    - 그래프객체.map(그래프 종류,기준)

```python
# <예제4-34> 조건에 맞게 화면 분할

import matplotlib.pyplot as plt
import seaborn as sns

titanic  = sns.load_dataset("titanic")

sns.set_style("whitegrid")

# 조건에 따라 그리드 나누기
g = sns.FacetGrid(data = titanic, col = "who", row = "survived")

#그래프 적용하기
g= g.map(plt.hist, "age")
```

이변수 데이터의 분포
- 열을 두 개씩 짝을 지어 모든 조합에 대하여 표시
    - 같은 것끼리는 히스토 그램
    - 다른 두개는 산점도
- pairplot() 함수사용

```python
# <예제 4-35> 이변수 데이터의 분포

import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset("titanic")

sns.set_style("whitegrid")

# titanic 데이터셋 중에서 분석 데이터 선택하기
titanic_pair = titanic[["age","pclass","fare"]]

#조건에따라그리드 나누기
g = sns.pairplot(titanic_pair)
```
