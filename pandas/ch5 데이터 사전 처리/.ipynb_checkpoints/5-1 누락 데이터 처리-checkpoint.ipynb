누락데이터
- NaN (Not a Number)
- 누락되는 경우
    - 데이터를 파일로 입력할 때 빠트리는 경우
    - 파일 형식을변환하는 과정
- 누락데이터 처리
    - 제거
    - 적절한 값으로 대체

누락 데이터 확인

- .isnull()
    - 누락 데이터 - True, 유효한데이터 -False 반환
- .notnull()
    - 누락 데이터 - False, 유효한데이터 -True 반환

```python
# <예제 5-1> 누락 데이터 확인

import seaborn as sns

# titanic 데이터 불러오기
titanic = sns.load_dataset("titanic")

print(titanic.head())

print("\n")

print(titanic.info())

print("\n")

# .value_counts(dropna = False 로 NaN 개수확인
print(titanic["deck"].value_counts(dropna = False))

print("\n")

# .isnull() 매소드로 누락 데이터 찾기
print(titanic.head().isnull())

print("\n")

# .isnull().sum(axis =0)으로 누락 데이터 개수 세기
print(titanic.head().isnull().sum(axis = 0))
```

누락 데이터 제거
- 열 삭제
    - 분석 대상이 갖는 특성(변수)를 제거
    - 데이터프레임 객체.dropna(axis = 1, thresh = n)
        - NaN의 수가 n개 이상 갖는 모든열을 삭제한다.

- 행 삭제
    - 분석 대상의 관측값(레코드)를 제거
    - 데이터프레임 객체.dropna(axis = 0, subset=["열 이름"], how = "any")
        - "열 이름"에 데이터가 없는 행 삭제
        - how = 옵션 
            - "any" (기본값) : NaN값이 하나라도 존재하면 삭제
            - "all" : 모든 데이터가 NaN값일 경우 삭제

```python
# <예제 5-2> 누락 데이터 제거
import seaborn as sns

df = sns.load_dataset("titanic")

# 열 삭제 (deck열 삭제)
df_thresh = df.dropna(axis= 1, thresh = 500) # NaN값이 500개 이상인 열을 삭제한다.
print(df_thresh)

print("\n")

# 행 삭제 ( age열의 데이터중 NaN값인 행을 삭제)
df_age = df.dropna(subset = ["age"], axis =0)
print(df_age)
```

누락 데이터 치환
- 데이터프레임 객체.fillna(대체할 값)
    - 원본객체 변경시 : inplace = True
- 대체할 값
    - 평균값, 최빈값, ...

```python
# <예제 5-3> 평균으로 누락 데이터 바꾸기

import seaborn as sns

df = sns.load_dataset("titanic")

# age 열의 첫 10 개 데이터 출력
print(df["age"].head(10))

print("\n")

# age열의 NaN갑을 다른 나이 데이터의 평균으로 변경하기
mean_age = df["age"].mean(axis = 0) # age 열의 평균 계산(NaN값 제외)
df["age"].fillna(mean_age,inplace =True)

print(df["age"].head(10))
```

```python
# <예제 5-4> 가장 많이 나타나는 값으로 바꾸기

import seaborn as sns

df =sns.load_dataset("titanic")

# embark_town 열의 829행의 NaN 데이터 출력
print(df["embark_town"][825:830])
print("\n")

# embark_town 열의 가장 많이 나오는 값 : most_freq
most_freq = df["embark_town"].value_counts(dropna = True).idxmax()
print(most_freq)
print("\n")

# embark_town 열의 NaN값을 most_freq로 치환
df["embark_town"].fillna(most_freq, inplace= True)

print(df["embark_town"][825:830])
```

이웃하고 있는 값으로 바꾸기
- fillna() 메소드에 옵션
    - method = "ffill"
        - 이전 값으로 바꿈
    - method= "bfill"
        - 다음 값으로 바꿈

```python
# <예제 5-5> 이웃하고 있는 값으로 바꾸기

import seaborn as sns

df = sns.load_dataset("titanic")

# embark_town 열의 829행의 NaN 데이터 출력
print(df["embark_town"][825:830])
print("\n")

# embark_town 열의 NaN값을 바로 앞에 있는 값으로 변경
df["embark_town"].fillna(method = "ffill", inplace = True)
print(df["embark_town"][825:830])
```
