단순 회귀 분석
- 두 변수 사이에 일대일로 대응되는 상관성을 찾는 알고리즘
- 대표적인 지도학습의 유형

- Y = aX + b
    - Y : 종속 변수(목표), X : 독립 변수
    - 기울기(a) 와 y절편(b)를 찾는 과정

```python
# <예제 7-1> 단순 회귀 분석

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

'''
step 1 데이터 준비 - read_csv() 함수로 자동차 연비 데이터 셋 가져오기
'''

df = pd.read_csv("./auto-mpg.csv",header = None)
df.columns = ["mpg","cylinders","displacement","horsepower","weight","acceleration","model year","origin","name"]

# Ipython 디스플레이 설정 - 출력할 열의 한도 개수 늘리기 4 -> 9
pd.set_option("display.max_columns",10)
print(df.head())

print("\n")

'''
step 2 데이터 탐색
df.info() 메소드 -> 데이터의 자료형과 개수
df.describe() 메소드 -> 주요 통계 정보
'''

print(df.info())
print("\n")

print(df.describe()) 
# "horsepower" 과 "name" 열은 문자열이므로 출력 X
# "horsepower" 열이 문자열 -> 실수형 변환 필요
print("\n")

# horsepower 열의 자료형 변경
print(df["horsepower"].unique()) # <- horsepower 열의 고유값 확인
print("\n")

df["horsepower"].replace("?",np.nan, inplace = True) # "?" -> np.nan 으로 변경
df.dropna(subset = ["horsepower"],axis = 0, inplace = True) # 누락 데이터 행 삭제
df["horsepower"] = df["horsepower"].astype("float") # 문자열 -> 실수형 변황

print(df.describe())
print("\n")

'''
step 3 속성 선택 
단순회귀분석에 변수로 사용할 후보 선택
'''

# 종속 변수 - "mpg" 열
# 독립 변수 후보 - "cylinders","horsepower","weight"

# 분석에 활용할 열 (속성) 선택
ndf = df[["mpg","cylinders","horsepower","weight"]]
print(ndf.head())
print("\n")

# 독립 변수 후보에서 독립 변수 선택하기
# - 종속변수와 독립변수 간의 선형관계가 있는지 확인 - 산점도 그래프

# seaborn pairplot()으로 두 변수 간의 모든 경우의 수 그리기
fig = plt.figure(figsize = (10,5))
grid_ndf = sns.pairplot(ndf)
plt.show()
plt.close() # ->"mpg" 와 선형관계를 나타내는 "horesepower" 와 "mpg"열을 독립변수로 선택

'''
step 4 한개의 독립변수를 -> 훈련/검증 데이터로 분할
'''

# "weight" 독립변수를 -> 훈련(274개) / 검증(118개)로 분할

# 속성(변수) 선택
y = ndf["mpg"]
X = ndf[["weight"]]

# 독립변수(weight) -> 훈련(train)데이터 와 검증(test)데이터 로 구분(7:3 비율)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, # 독립변수
                                                   y, # 종속변수
                                                   test_size = 0.3, # 검증 30%
                                                   random_state =10) # 랜덤 추출값

print("train data 개수: ", len(X_train))
print("test data 개수: ", len(X_test))

print("\n")

'''
step 5 모형 학습 및 검증
1) LinearRegression() -> 회귀분석 모형 객체 생성 -> 변수 lr에 저장
2) 모형객체(lr).fit(훈련 데이터(X_train,y_train)) -> 목표값(a,b)를 찾음
3) lr.score(X_test,y_test) -> 결정 계수를 구한다.(클수록 예측 능력이 좋다.) -> r_square변수에 저장
'''

# Sklearn 라이브러리에서 선형회귀분석 모듈 가져오기
from sklearn.linear_model import LinearRegression

# 단순회귀 분석 모형 객체 생성
lr = LinearRegression()
# train data를 가지고 모형 학습
lr.fit(X_train, y_train)
# 결정계수 계산
r_square =lr.score(X_test,y_test)
print(r_square)

print("\n")

# 목표값 출력
# 회귀식의 기울기 - a
print("기울기 a:", lr.coef_)
# 회귀식의 y절편 - b
print("y절편 b:", lr.intercept_)

# 모형이 예측한 결과 <-> 실제 결과값 비교
# y_hat(예측한 결과) = lr.predict(X(독립변수))
# y_hat 과 y(실제 결과값) -> 같은 화면에 그리기
y_hat = lr.predict(X)

plt.figure(figsize = (10,5))
ax1 = sns.distplot(y, hist = False, label = "y")
ax2 = sns.distplot(y_hat, hist = False, label = "y_hat", ax =ax1)
plt.show()
plt.close()
```

결과
- 실제값(y) - 왼쪽으로 편향
- 예측값(y_hat) - 오른쪽으로 편향

- 오차를 더 줄여야 한다.
